```python
import numpy as np
```

```python
#creating array
arr1 = np.array([1,2,3])   #-->1d
arr2 = np.array([[1,2,3],[4,5,6]])   #-->2d
print(arr1)
print()
print(arr2)
```

```python
print(arr1.shape)
print(arr1.size)
print(arr1.ndim)
print()
print(arr2.shape)
print(arr2.size)
print(arr2.ndim)
```

```python
#zeroes ones and random (weights /dummy data)

zeroes = np.zeros((3,2))
ones = np.ones((4,2))
random1 = np.random.rand(4,3)
random2 = np.random.rand(5,3)

print(zeroes)
print()
print(ones)
print()
print(random1)
print()
print(random2)
```

```python
# indexing and slicing (feature selection)
arr2[:,1]
```

```python
#vectorized operations for fast ml
print(arr2 +2)
print(arr2 *2)
print(arr2 /2)
```

```python
#data scaling
print(np.mean(arr2))
print()
print(np.std(arr2))
print()
print(np.sum(arr2))
```

```python
#reshaping the entrire array
old_arr = np.array([[5,4,3],[1,2,3]])
print(old_arr)
new_arr = old_arr.reshape(3,2)
print(new_arr)
```

```python
#dot product
X= np.array([[1,2,3],[4,5,6]])   #-->2d
Y= np.array([[1,2,3],[4,5]])   #-->2d
np.dot(X,Y)
```
